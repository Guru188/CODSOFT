```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = {
    'title': [
        'Avatar', 'Titanic', 'Avengers', 'Iron Man',
        'The Dark Knight', 'Interstellar', 'Inception'
    ],
    'genres': [
        'Action Adventure Sci-Fi',
        'Romance Drama',
        'Action Adventure Sci-Fi',
        'Action Sci-Fi',
        'Action Crime Drama',
        'Sci-Fi Adventure Drama',
        'Action Sci-Fi Thriller'
    ]
}

df = pd.DataFrame(data)

vectorizer = CountVectorizer()
genre_matrix = vectorizer.fit_transform(df['genres'])

similarity_matrix = cosine_similarity(genre_matrix)

def recommend_movie(movie_name):
    movie_name = movie_name.title()

    if movie_name not in df['title'].values:
        print("Movie not found in database")
        print("Available movies:")
        for movie in df['title']:
            print(movie)
        return

    index = df[df['title'] == movie_name].index[0]
    scores = list(enumerate(similarity_matrix[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("Recommended movies similar to", movie_name)
    count = 0
    for i in scores:
        if df.iloc[i[0]]['title'] != movie_name:
            print(df.iloc[i[0]]['title'])
            count += 1
        if count == 3:
            break

user_movie = input("Enter a movie name: ")
recommend_movie(user_movie)
```
